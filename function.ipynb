```python
def c_to_k(temperature):
    temperature = temperature + 273.15
    return temperature
```

```python
print(c_to_k(50.0))
```

```python
def f_to_c(temperature):
    temperature = (temperature-32)*5/9
    return temperature
```

```python
print(f_to_c(-40.0))
```

```python
def f_to_k(temperature):
    temperature = c_to_k(f_to_c(temperature))
    return temperature
```

```python
print(f_to_k(32))
```

```python
def check_data(data):
    if (data.max(axis=0)[0]==0.0) and (data.max(axis=0)[20]==20.0):
        print("Odd max")
    elif (sum(data.min(axis=0)) == 0.0):
        print("Odd min")
    else:
        print("OK")
```

```python
def plot_data(data):
    fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))
    axis1 = fig.add_subplot(1,3,1)
    axis2 = fig.add_subplot(1,3,2)
    axis3 = fig.add_subplot(1,3,3)
    
    axis1.set_ylabel("Mean")
    axis1.plot(data.mean(axis=0))
    
    axis2.set_ylabel("Max")
    axis2.plot(data.max(axis=0))
    
    axis3.set_ylabel("Min")
    axis3.plot(data.min(axis=0))
    
    fig.tight_layout()
    matplotlib.pyplot.show(fig)
```

```python
%matplotlib inline
import matplotlib.pyplot
import numpy
```

```python
import glob
```

```python
files = glob.glob('data/temperature-*.csv')
```

```python
print(files)
```

```python
for file in files:
    print(file)
    data = numpy.loadtxt(fname=file, delimiter=',')
    check_data(data)
    plot_data(data)
```
